```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch
from torchmetrics import AUROC
import torch.nn.functional as F
from scipy import stats

from func2graph import models, tools, baselines, data
```

```python
v = np.load('../../data/Ziyu/200_bio_sim/sim_voltage.npy')
print(v.shape)

plt.figure(figsize=(5, 8))
for i in range(20):
    plt.subplot(20, 1, i+1)
    plt.plot(v[i,:][:200])
plt.show()

# Clip to 30 and normalize the data for each neuron
v[v > 30] = 30
v = (v - np.mean(v, axis=1, keepdims=True)) / np.std(v, axis=1, keepdims=True)

plt.figure(figsize=(5, 8))
for i in range(20):
    plt.subplot(20, 1, i+1)
    plt.plot(v[i,:][:200])
plt.show()

print("new 200-neuron data")
print('max:', np.max(v))
print('min:', np.min(v))
print('mean:', np.mean(v))
```
